```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve, precision_recall_curve
```

```python
train_feature= pd.read_csv('/content/training_set_features.csv')
train_feature.head()
```

```python
train_label=pd.read_csv('/content/training_set_labels.csv')
```

```python
train_label.head()
```

```python
test_feature=pd.read_csv('/content/test_set_features.csv')
```

```python
test_feature.head()
```

```python
merged_train_feature=pd.merge(train_feature,train_label,on='respondent_id')
```

```python
merged_train_feature.head()
```

```python
train=merged_train_feature.drop(['respondent_id','xyz_vaccine','seasonal_vaccine'],axis=1)
train.head()
```

```python
test=test_feature.drop(['respondent_id'],axis=1)
test.head()
```

```python
train.info()
```

```python
test.info()
```

```python
train.select_dtypes(include=['object']).columns.tolist()
numerical_features = train.select_dtypes(include=[np.number]).columns.tolist()
```

```python
categorical_features = train.select_dtypes(include=['object']).columns.tolist()
numerical_features = train.select_dtypes(include=[np.number]).columns.tolist()
```

```python
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier())
])
```

```python
y_xyz=merged_train_feature['xyz_vaccine']
```

```python
y_seas=merged_train_feature['seasonal_vaccine']
```

```python
X_train, X_val, y_train_xyz, y_val_xyz, y_train_seasonal, y_val_seasonal = train_test_split(train, y_xyz, y_seas, test_size=0.2, random_state=42)
```

```python
xyz_vaccine=model.fit(X_train, y_train_xyz)
y_pred_xyz = model.predict_proba(X_val)[:, 1]
roc_auc_xyz = roc_auc_score(y_val_xyz, y_pred_xyz)
```

```python
print(f'ROC AUC for xyz_vaccine: {roc_auc_xyz}')
```

```python
seasonal_vaccine= model.fit(X_train, y_train_seasonal)
y_pred_seasonal = model.predict_proba(X_val)[:, 1]
roc_auc_seasonal = roc_auc_score(y_val_seasonal, y_pred_seasonal)
```

```python
print(f'ROC AUC for seasonal_vaccine: {roc_auc_seasonal}')
```

```python
model.fit(train, y_xyz)
test_pred_xyz = model.predict_proba(test)[:, 1]

model.fit(train, y_seas)
test_pred_seasonal = model.predict_proba(test)[:, 1]
```

```python
submission = pd.DataFrame({
    'respondent_id': test_feature['respondent_id'],
    'xyz_vaccine': test_pred_xyz,
    'seasonal_vaccine': test_pred_seasonal
})
```

```python
submission.to_csv('submission.csv',index=False)
```
